```python
import torchvision
from torchvision import transforms
from torch.utils.data import DataLoader
from matplotlib.pyplot import imshow
%matplotlib inline
```

```python
train_data = torchvision.datasets.ImageFolder(root = "dog", transform = None)
```

```python
for data,label in train_data:
    imshow(data)
    break
```

```python
trans = transforms.Compose([
    transforms.Resize((64, 128))
])
train_data = torchvision.datasets.ImageFolder(root = "dog", transform = trans)
```

```python
for data,label in train_data:
    imshow(data)
    break
```
